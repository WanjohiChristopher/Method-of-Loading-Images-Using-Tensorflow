```python
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filename='dog.jpeg'
```

# Methods of loading images

```python
# first method
from IPython.display import Image
Image(filename='dog.jpeg',width=224,height=224)
```

```python
#second method
from tensorflow.keras.preprocessing import image
img=image.load_img(filename,target_size=(224,224))
```

```python
plt.imshow(img)
```

```python
#third method
import cv2
imd=cv2.imread(filename)
plt.imshow(imd)
```

```python
imd=cv2.resize(imd,(224,224))
plt.imshow(imd)
```

```python
#forth method
from PIL import  Image
im=Image.open(filename)
im=im.resize((224,224))
plt.imshow(im)
```

# Deeplearning Models

```python
#mobile=tf.keras.applications.mobilenet.MobileNet()# version 1
mobile=tf.keras.applications.mobilenet_v2.MobileNetV2()# versio 2
#mobile=tf.keras.applications.ResNet152V2.ResNet152V2()              # version 3
```

## Image preprocessing

```python
from tensorflow.keras.preprocessing import image
img=image.load_img(filename,target_size=(224,224))
plt.imshow(img)
```

```python
#resizing
import numpy as np

resized_img=image.img_to_array(img)
final_image=np.expand_dims(resized_img,axis=0)# we require forth dimension
final_image=tf.keras.applications.mobilenet.preprocess_input(final_image)
```

```python
final_image.shape
```

```python
predictions=mobile.predict(final_image)
```

```python
from tensorflow.keras.applications import  imagenet_utils

results=imagenet_utils.decode_predictions(predictions)
```

```python
print(results)
```
